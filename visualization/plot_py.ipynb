```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips_xxx = tips.groupby(['time']).agg({'total_bill' : 'sum', 'tip' : 'sum'}).apply(lambda x: 100 * x / float(x.sum()))
```

```python
tips_xxx
```

```python
tips_xxx.T.plot.bar(stacked=True, figsize=(12,6))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'smoker', y='total_bill', hue='sex', data = tips)
plt.title('Smoker Distribution')
plt.ylabel('Total Bill')
plt.xlabel('Smoker or Not')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.loc[flights['month'] == 'May']
```

```python
flights.query("month == 'May'")
```

```python
may_flights = flights.query("month == 'May'")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=may_flights, x="year", y="passengers")
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()
```

```python
flights_wide_number = flights.pivot("year", "month", "passengers")
```

```python
flights.pivot("year", "month", "passengers")
```

```python
flights_wide = flights.pivot("year", "month", "passengers").apply(lambda x: 100 * x / float(x.sum()))
```

```python
flights_wide
```

```python
flights_wide.T.plot.bar(stacked=True, figsize=(12,6))
```

```python
flights_wide_number.plot.area()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=flights_wide)
```
