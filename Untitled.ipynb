```python
def calculate_age(born):
    born = datetime.strptime(born, "%Y-%m-%d")
    today = date.today()
    age = today.year - born.year - ((today.month, today.day) < (born.month, born.day))
    return age
```

```python
# function / lambda

df_3["age"] = df_3["dt_date_of_birth"].apply(lambda x: calculate_age(x))
```

```python
def add_five(value):
    value = value + 5
    return value
```

```python
def add_five_multiple(value1, value2):
    value = value1 + value2 + 5
    return value
```

```python
def add_five_multiple_2(value1, value2):
    value_a = value1 + value2 + 5
    value_b = value1 + value2 + 10
    return value_a, value_b
```
